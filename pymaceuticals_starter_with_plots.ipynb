```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
```

```python
# Combine the data into a single dataset
df = pd.merge(study_results, mouse_metadata, how='left', left_on='Mouse ID', right_on='Mouse ID')
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
regimen = {'Capomulin' : [], 'Ramicane': [], 'Infubinol': [],  'Ceftamin': []}
ogkeys = list(regimen.keys())
mice = df['Mouse ID'].value_counts()
df2 = pd.DataFrame(mice).reset_index()
df2.columns = ['Mouse ID','Count']
df_dedup2 = df_dedup.set_index('Mouse ID')
max_timepoint = []
id_mouse = []
size_tumor = []
for mice in df_dedup2.index.unique():
    max_timepoint.append(df_dedup2.loc[mice,'Timepoint'].max())
    id_mouse.append(mice)
    
# Start by getting the last (greatest) timepoint for each mouse
a = {'id_mouse': id_mouse, 'max_timepoint': max_timepoint}
tumor_df = pd.DataFrame(a)
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
new_tumor_df = pd.merge(tumor_df, df_dedup,  how='left', left_on=['id_mouse', 'max_timepoint'], right_on =['Mouse ID', 'Timepoint'])
#da = tumor_df['id_mouse'] + str(tumor_df['max_timepoint'])
#tumor_df['unique_id'] = da
#da
quartile_df = new_tumor_df.drop(columns=['Mouse ID', 'Timepoint'] )
quartile_df
```

```python
#listreg = list(regimen)
for key in list(regimen.keys()):
    i = str(key)
    i = quartile_df.loc[quartile_df['Drug Regimen'] == key,'Tumor Volume (mm3)'].tolist()
    regimen[key] = i
    
    
```

```python
for key, value in list(regimen.items()):
    j = pd.Series(value, dtype='float64')
    quartiles = j.quantile([.25,.5,.75])
    lowerq = quartiles[.25]
    upperq = quartiles[.75]
    iqr = upperq-lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    regimen['upperq_%s' % key] = upperq
    regimen['lowerq_%s' % key] = lowerq
    regimen['iqr_%s' % key] = iqr
    regimen['lbound_%s' % key] = lower_bound
    regimen['ubound_%s' % key] = upper_bound
```

```python
for key, v in list(regimen.items()):
    outlier = []
    if key in ogkeys:
        for m in v:
            if m < regimen['lbound_%s' % key] or m > regimen['ubound_%s' % key]:
                outlier.append(m)
        regimen['outliers_%s' % key] = outlier
    
        
```
